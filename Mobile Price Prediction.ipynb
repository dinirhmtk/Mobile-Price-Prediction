# Introduction

For this project, the mobile price dataset used from <a href="https://www.kaggle.com/c/sanbercode-ds-final-project/data">Kaggle</a>. The data contains the following fields:

- index - id unique number which has an integer positive value
- daya_baterai - total energy stored in the battery (in mAh)
- bluetooth - has bluetooth (1: yes; 0: no)
- kecepatan_clock - processor speed when executes the instructions
- dual_sim - has dual sim capability (1: yes; 0: no)
- kamera_depan - large resolution of front camera (in MegaPixels)
- four_g - has 4G connectivity (1: yes; 0: no)
- memori_internal - large internal memory (in GB)
- tebal_hp - mobile thickness (in cm)
- berat_hp - mobile mass (in gram)
- jumlah_prosesor - number of cores in the processors (pcs)
- kamera_belakang_mp - large resolution of rear camera (in MegaPixels)
- px_panjang - length of the pixel resolution
- px_lebar - width of the pixel resolution
- kapasitas_ram - capacity of Random Access Memory (in MB)
- panjang_layar - screen length (satuan cm)
- lebar_layar - screen width (satuan cm)
- waktu_telepon - the longest call time that can be made from a fully charged mobile until it runs out (in hours)
- three_g - has 3G connectivity (1: yes; 0: no)
- touch_screen - has touch screen (1: yes; 0: no)
- wifi - has wifi (1: yes; 0: no)
- price_range - mobile range price (0-3000)

3. Data preprocessing and data cleansing

Process contained at this step is removing some missing values, encoding the data to change categorical into numerical data type and drop some unnecessary columns. The data still have missing values that can cause inaccurate analysis. This step below show the process of removing some missing values with mean of the column using fillna() methods

```python
train['kecepatan_clock'].fillna(train['kecepatan_clock'].mean(), inplace=True)
train['panjang_layar'].fillna(train['panjang_layar'].mean(), inplace=True)
```

Changing categorical type to numerical type using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

le_four_g = LabelEncoder()
# encoding untuk kolom four_g
le_four_g.fit(train['four_g'])
encoded_four_g_train = le_four_g.transform(train['four_g'])
train['four_g'] = encoded_four_g_train

# encoding untuk kolom price_range
le_four_g.fit(train['price_range'])
encoded_four_g_train = le_four_g.transform(train['price_range'])
train['price_range'] = encoded_four_g_train

train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_four_g = LabelEncoder()
# encoding untuk kolom four_g
le_four_g.fit(test['four_g'])
encoded_four_g_test = le_four_g.transform(test['four_g'])
test['four_g'] = encoded_four_g_test

test.head()
```

```python
train.corr()
```

Droping unnecessary columns based on correlation rate. Columns that I chose are the one which has a correlation rate with price_range column less than 0,005 whether it negative or positive because negative correlation value means that the relationship between the data is reversed.

```python
drop_element_train = ['index','panjang_layar','kecepatan_clock','wifi']
drop_element_test = ['panjang_layar','kecepatan_clock','wifi']

train = train.drop(drop_element_train, axis=1)
test = test.drop(drop_element_test, axis=1)
```

```python
train = train.astype(int)
test = test.astype(int)
```

5. Model selection and evaluation

Algorithm model that used below is KNN Classifier with some metrics and RandomizedSearchCV for model selection. At this step, it can be seen some aspects that are best for predicting data like, best parameters, best scores, and best estimators. After finding the best parameters, the next step is ask data to learn the model using fit() function.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import RandomizedSearchCV
```

Process of separating feature data and target data

```python
x = train.drop('price_range', axis=1)
y = train['price_range']
```

Turning hyperparameter process with KNN algorithm and then make the dataset learn the model. You can see the complete KNeighbours Classifier in this <a href="https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html">link</a>. I used n_neighbors params with range 2 to 50.

```python
model_knn = KNeighborsClassifier()
param_grid = {'n_neighbors':np.arange(2,50),
              'weights':['uniform','distance']}
rscv = RandomizedSearchCV(model_knn, param_grid, cv=10, scoring='f1_micro')
rscv.fit(x,y)
```

```python
rscv.best_params_
```

```python
rscv.best_score_
```

```python
rscv.best_estimator_
```

```python
model_knn = KNeighborsClassifier(n_neighbors=38, weights='distance')
model_knn.fit(x,y)
```

```python
fitur_train = train[['daya_baterai','bluetooth','dual_sim','kamera_depan','four_g','memori_internal',
                     'tebal_hp','berat_hp','jumlah_prosesor','kamera_belakang_mp','px_panjang','px_lebar',
                     'kapasitas_ram','lebar_layar','waktu_telfon','three_g','touch_screen']]

fitur_test = test[['daya_baterai','bluetooth','dual_sim','kamera_depan','four_g','memori_internal',
                   'tebal_hp','berat_hp','jumlah_prosesor','kamera_belakang_mp','px_panjang','px_lebar',
                   'kapasitas_ram','lebar_layar','waktu_telfon','three_g','touch_screen']]

x = np.array(fitur_train)
y = train['price_range']
x_test = np.array(fitur_test)
```

Predict on the model data for evaluation to know the fmean score (f1_score) using data train.

```python
y_pred_train = model_knn.predict(x)
```

```python
y_pred_train.shape
```

```python
f1_score(y, y_pred_train, average='weighted')
```

Predict on the model data for submission file using data test.

```python
y_pred_test = model_knn.predict(x_test)
```

```python
y_pred_test.shape
```

Make a submission file of mobile price prediction

```python
submission = pd.DataFrame({'index':test['index'],
                           'price_range':y_pred_test})
submission.to_csv('result.csv', index = False)
```
